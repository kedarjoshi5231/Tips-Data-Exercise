```python
# Setting up Seaborn
sns.set(style="whitegrid")
tips = sns.load_dataset("tips")
```

```python
ax = sns.boxplot(x="day", y="tip", data=tips)
```

###### Using the graph, we can compare the range and distribution of the area_mean for the tips. We observe that interquartile range (IQR) is close enough amongst thursday ,friday and saturday. Whereas it increases by small amount on sunday. Even if range of tips(mean area) on friday is lesser than sunday, people on an average tip(median) almost similar to people on sunday. Average tips on thursday are significantly lesser concluding people visiting on thursday aren't keen to give tips for the service provided. But, there are some oultliers which tend to pay more than maximum on thursdays. Suprisingly, when you compare saturdays and sundays, the range of maximum and minimum is lesser on saturday but there are more outliners which means the customer satisfaction might be higher on sundays but some people on saturdays are willing to pay more than expected on saturdays.

```python
ax = sns.boxplot(x="time", y="total_bill",hue="smoker", data=tips, palette="Set2")
```

###### If we observe here, smokers tend to spend less than non-smokers on lunch but its vice versa when it comes to dinner. The medians of both smokers and non smokers tend to match for their lunch expense but the IQR is wider and maximum is significantly higher for non smokers for lunch. On the other side we see, the average total bill of smokers is higher compared to non-smokers but lesser outliers

```python
# create dummy variable then group by that
tips.assign(dummy = 1).groupby(['dummy','sex']).size().groupby(level=0).apply(lambda x: 100 * x / x.sum()).to_frame().unstack().plot(kind='bar',legend=False)

# or it'll show up as 'dummy' 
plt.xlabel('Gender')

# disable ticks in the x axis
plt.xticks([])

# fix the legend or it'll include the dummy variable
current_handles, _ = plt.gca().get_legend_handles_labels()
reversed_handles = reversed(current_handles)
correct_labels = reversed(tips['sex'].unique())

plt.legend(reversed_handles,correct_labels)

plt.gca().yaxis.set_major_formatter(mtick.PercentFormatter())
plt.show()
```
